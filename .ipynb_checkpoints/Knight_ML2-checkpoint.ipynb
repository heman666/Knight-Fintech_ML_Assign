```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('train.csv')
print(len(set(data.variety)))
data.head()
```

```python
print(len(set(data.variety.values)))
data.drop(['user_name','review_description','region_2'],axis='columns',inplace=True)
data.head()
```

```python
data = data[data['country'].notna()]
data = data[data['province'].notna()]
data = data[data['region_1'].notna()]
data = data[data['winery'].notna()]
data = data[data['review_title'].notna()]
data = data[data['designation'].notna()]
data = data[data['points'].notna()]
data = data[data['price'].notna()]
# data = data[data['region_2'].notna()]
data = data[data['variety'].notna()]
print(data.shape)
print(data.info())
print(len(set(data.variety.values)))
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

X = data.iloc[:,:-1]
variety = pd.get_dummies(data.variety)
print(len(variety))
data = pd.concat([data,variety],axis='columns') 
data.drop(['variety','Syrah'],axis='columns',inplace=True)
x_train = X.values
Y = data.iloc[:,9:]
y_train = Y.values

# data.head()
Y
```
