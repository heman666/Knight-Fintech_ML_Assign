```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from catboost import Pool, CatBoostClassifier, cv
```

```python
data = pd.read_csv('train.csv')
print(data.shape)
print(data.info())
data.head()
```

```python
data.drop(['user_name','review_description','region_2','region_1'],axis='columns',inplace=True)
data.head()
```

```python
data = data[data['country'].notna()]
data = data[data['province'].notna()]
# data = data[data['region_1'].notna()]
data = data[data['winery'].notna()]
data = data[data['review_title'].notna()]
data = data[data['designation'].notna()]
data = data[data['points'].notna()]
data = data[data['price'].notna()]
data = data[data['variety'].notna()]
print(data.shape)
print(data.info())
print(len(set(data.variety.values)))
data.head()
```

```python
x = data.iloc[:,:-1]
y = data.iloc[:,-1]
categorical_features_indices = np.where(x.dtypes != np.float)[0]
int_data = np.where(data.dtypes == np.int)[0]
# categorical_features_indices = categorical_features_indices[0:3] + categorical_features_indices[4:]
categorical_features_indices = np.delete(categorical_features_indices,int_data)
print(categorical_features_indices)
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1)
model = CatBoostClassifier(iterations=300, learning_rate=0.01, l2_leaf_reg=3.5, depth=8, rsm=0.98, loss_function= 'MultiClass', eval_metric='AUC',use_best_model=True)

print(x_train.shape)
print(y_train.shape)
model.fit(x_train,y_train,cat_features=categorical_features_indices,eval_set=(x_test,y_test))
```
